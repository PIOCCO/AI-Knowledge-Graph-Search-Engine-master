```python
# Core
import pandas as pd
import numpy as np
from datetime import datetime

# ML
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report, accuracy_score
from sklearn.preprocessing import LabelEncoder

# Persistence
import joblib

print("All imports loaded successfully")

# Load data
df = pd.read_csv("Ticket.csv")
print(df.head())
print("\nCategory column type:", df["category"].dtype)
print("Unique categories:", df["category"].unique())

# ✅ FIX: Convert categories to integers using LabelEncoder
label_encoder = LabelEncoder()
df["category_encoded"] = label_encoder.fit_transform(df["category"])

# Save the label encoder for later use
joblib.dump(label_encoder, "../label_encoder.pkl")

print("\nCategory mapping:")
for idx, label in enumerate(label_encoder.classes_):
    print(f"  {idx}: {label}")

# Create text features
df["text"] = df["title"] + " " + df["description"]
X = df["text"]
y = df["category_encoded"]  # ✅ Use encoded integers

# Split data
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.4, random_state=42, stratify=y
)

# Vectorize
vectorizer = TfidfVectorizer(
    stop_words="english",
    max_features=5000,
    ngram_range=(1, 2)
)
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)

# Train model
model = MultinomialNB()
model.fit(X_train_vec, y_train)

# Evaluate
y_pred = model.predict(X_test_vec)
print("\n✅ Accuracy:", accuracy_score(y_test, y_pred))
print("\n" + classification_report(y_test, y_pred, target_names=label_encoder.classes_))

# Save everything
joblib.dump(model, "../ml_service/ticket_classifier.pkl")
joblib.dump(vectorizer, "../ml_service/tfidf_vectorizer.pkl")
print("\n✅ Model, vectorizer, and label encoder saved")
```
